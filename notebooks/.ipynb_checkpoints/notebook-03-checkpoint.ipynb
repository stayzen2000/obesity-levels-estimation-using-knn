```python
import pandas as pd

X = pd.read_csv(r"C:\Users\aleaf\OneDrive\Desktop\Projects\obesity-levels-estimation-using-knn\data\processed\X_features.csv")
y = pd.read_csv(r"C:\Users\aleaf\OneDrive\Desktop\Projects\obesity-levels-estimation-using-knn\data\processed\y_target.csv").squeeze()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_euclidean = KNeighborsClassifier(n_neighbors=5, metric="euclidean")
knn_manhattan = KNeighborsClassifier(n_neighbors=5, metric="manhattan")

knn_euclidean.fit(X_train, y_train)
knn_manhattan.fit(X_train, y_train)
```

```python
print("Euclidean Train:", knn_euclidean.score(X_train, y_train))
print("Euclidean Test:", knn_euclidean.score(X_test, y_test))

print("Manhattan Train:", knn_manhattan.score(X_train, y_train))
print("Manhattan Test:", knn_manhattan.score(X_test, y_test))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances

eu_dist = pairwise_distances(X_train, metric="euclidean").flatten()
man_dist = pairwise_distances(X_train, metric="manhattan").flatten()

plt.hist(eu_dist, bins=50, alpha=0.5, label="Euclidean")
plt.hist(man_dist, bins=50, alpha=0.5, label="Manhattan")
plt.legend()
plt.title("Distance Distributions (Lifestyle Features)")
plt.show()
```
